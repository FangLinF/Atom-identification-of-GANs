```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def judge_atomPos(imagePath,fixdis_value,pixel_threshold,save_txt_path):
    
    meta_image = Image.open(imagePath)
    plt.imshow(meta_image)

    gray_image = meta_image.convert('L')
    plt.imshow(gray_image,'gray')

    metaImage_data = np.array(meta_image)
    grayImage_data = np.array(gray_image)

    [sx ,sy] = np.where(grayImage_data > pixel_threshold)
    sxLen = len(sx)
    sx = sx.reshape((1,sxLen))
    print(sx,sx.shape,sx.dtype)
    newxrow = np.tile(sx,[sxLen,1])
    print(newxrow,newxrow.shape,newxrow.dtype)
    newxcolum = np.tile(sx.T,[1,sxLen])
    print(newxcolum,newxcolum.shape,newxcolum.dtype)
    dissx = abs(newxcolum - newxrow)
    print(dissx,dissx.shape,dissx.dtype)
    flagsx = (dissx < fixdis_value)
    print(flagsx,flagsx.shape,flagsx.dtype)
    coorsx = np.zeros((sxLen,sxLen),dtype = 'int64')
    print(coorsx,coorsx.shape,coorsx.dtype)
    for i in range(sxLen):
        [temp_index] = np.where(flagsx[i,:] == True)
        coorsx[i,0:len(temp_index)] = temp_index
    print(coorsx,coorsx.shape,coorsx.dtype)
    print('********************************************************************')
    syLen = len(sy)
    sy = sy.reshape((1,syLen))
    print(sy,sy.shape,sy.dtype)
    newyrow = np.tile(sy,[syLen,1])
    print(newyrow,newyrow.shape,newyrow.dtype)
    newycolum = np.tile(sy.T,[1,syLen])
    print(newycolum,newycolum.shape,newycolum.dtype)
    dissy = abs(newycolum - newyrow)
    print(dissy,dissy.shape,dissy.dtype)
    flagsy = (dissy < fixdis_value)
    print(flagsy,flagsy.shape,flagsy.dtype)
    coorsy = np.zeros((syLen,syLen),dtype = 'int64')
    print(coorsy,coorsy.shape,coorsy.dtype)
    for i in range(syLen):
        [temp_index] = np.where(flagsy[i,:] == True)
        coorsy[i,0:len(temp_index)] = temp_index
    print(coorsy,coorsy.shape,coorsy.dtype)

    fenlei = np.zeros((sxLen,(fixdis_value+2)*(fixdis_value+2)),dtype='int64')
    print(fenlei,fenlei.shape,fenlei.dtype)
    for i in range(fenlei.shape[0]):
        mtemp = np.intersect1d(coorsx[i,:],coorsy[i,:])
        fenlei[i,0:len(mtemp)] = mtemp
    print(fenlei,fenlei.shape,fenlei.dtype)

    atomPos = np.zeros((sxLen,2))
    print(atomPos,atomPos.shape,atomPos.dtype)
    len_fenlei = len(fenlei[0,:])
    print(len_fenlei)
    for i in range(sxLen):
        tempm = fenlei[i,:];
        temp_list = []
        for j in range(len_fenlei):
            if fenlei[i,j] != 0:
                temp_list.append(fenlei[i,j])
        if len(temp_list) == 0:
            temp_list.append(0)
        temp_list = np.array(temp_list,dtype='int64')
    #     print('**********')
    #     print(temp_list,temp_list.shape,temp_list.dtype)
    #     print(sy[0,temp_list] + 1)
    #     print(sx[0,temp_list] + 1)
    #     print('**********')
        atomX = np.mean(sy[0,temp_list] + 1)
        atomY = np.mean(sx[0,temp_list] + 1)
        atomPos[i,:] = [atomX,atomY]

    savetxtpath = save_txt_path + imagePath.split('/')[-1].split('.')[0] + '.txt'
    np.savetxt(savetxtpath,atomPos) 
```
